```python
import random
import sys

def PlayingSudoku():                             # 스도쿠를 실행시키는 함수
    answer = makeSudokuAnswer()
    quiz = makeSudokuQuiz(answer)
    current=[4,4]
    
    while(1):
        printSudoku(quiz, current)
        checkingAnswer2(answer, quiz, current)
        if insertkeypad(current, quiz)==1:
            print("""
        ============================
        
               MISSION FAILED
        
        ============================
        """)
            break


def makeSudokuAnswer():                           # 스도쿠 정답를 만드는 함수
    answer = makeSudokuCase()
    fillSudoku(answer)
    for i in range(100):
        mixSudoku(answer)
    return answer

def makeSudokuQuiz(answer):                      # 스도쿠 문제지를 만드는 함수
    quiz = makeSudokuCase()
    copySudoku(answer, quiz)
    makeBlankSudoku(quiz)
    return quiz

def checkingAnswer(Answer, Quiz):                # 정답 체크함수
    result = 0
    for i in range(9):
        for j in range(9): 
            if (Answer[i][j] != abs(int(Quiz[i][j]))) or Answer[i][j] == "0":
                result = 1
                break
        if (result == 1):
            break
        print("""
        ============================
        
                CONGRATULATION
               MISSION COMPLETE
        
        ============================
        """)
    return result

def checkingAnswer2(Answer, Quiz, current):      # 스도쿠 1개만 비교하는함수
    if Answer[current[0]][current[1]] == Quiz[current[0]][current[1]]:
        Quiz[current[0]][current[1]] = abs(Quiz[current[0]][current[1]])
        print("맞음")
        


def makeSudokuCase():                            # 스도쿠 빈 케이스을 만들어주는 함수
    result=[['a' for i in range(9)] for i in range(9)]
    return result


def randomIntList(num1, num2, num3, list1):     # 중복이 안되게 (num1~num2)중 (num3)개를 랜덤으로 리스트에 넣어주는 함수
    for i in range(num3):
        a = random.randint(num1, num2)
        while a in list1:
            a = random.randint(num1, num2)
        list1.append(a)

def randomTwo(num1, num2):                      # num1~num2 중 무작위 2개 뽑아서 반환하는 함수
    result1 = random.randint(num1, num2)
    result2 = random.randint(num1, num2)
    while result1 == result2:
        result2 = random.randint(num1,num2)
    return result1, result2
        
def ov(num):                                    # 오버플로우 함수
    if num > 8:
        result = num - 9
    else:
        result = num
    return result


def fillSudoku(sudoku):                         # 만들어진 스도쿠틀에 숫자를 넣는 함수
    list1=[]
    randomIntList(1, 9, 9, list1)
    
    for i in range(9):
        a=ov(i+3)
        b=ov(i+6)
        sudoku[0][i] = list1[i]
        sudoku[1][i] = list1[a]
        sudoku[2][i] = list1[b]
        
    for i in range(9):
        a = ov(i+1)
        sudoku[3][i] = sudoku[0][a]
        sudoku[4][i] = sudoku[1][a]
        sudoku[5][i] = sudoku[2][a]
    
    for i in range(9):
        a=ov(i+1)
        sudoku[6][i] = sudoku[3][a]
        sudoku[7][i] = sudoku[4][a]
        sudoku[8][i] = sudoku[5][a]



def mixSudoku(sudoku):                     # 스도쿠를 섞어주는 함수를 무작위로 고르는 함수
    num = random.randint(1,4)
    if num==1:
        mixSudoku1(sudoku)
    elif num==2:
        mixSudoku2(sudoku)
    elif num==3:
        mixSudoku3(sudoku)
    elif num==4:
        mixSudoku4(sudoku)
    
def mixSudoku1(sudoku):                    # 같은 묶음 안에서의 가로줄 교환
    num = random.randint(0,2)*3
    a, b = randomTwo(num, num+2)
    list1=[]
    for i in range(9):
        list1.append(sudoku[a][i])
    for i in range(9):
        sudoku[a][i]=sudoku[b][i]
    for i in range(9):
        sudoku[b][i]=list1[i]
    
def mixSudoku2(sudoku):                    # 같은 묶음 안에서의 세로줄 교환
    num = random.randint(0,2)*3
    a, b = randomTwo(num, num+2)
    list1=[]
    for i in range(9):
        list1.append(sudoku[i][a])
    for i in range(9):
        sudoku[i][a]=sudoku[i][b]
    for i in range(9):
        sudoku[i][b]=list1[i]
        
def mixSudoku3(sudoku):                     # 서로 다른 묶음 끼리의 가로줄 교환
    a, b = map(lambda x:x*3,randomTwo(0,2))
    list1=[[] for i in range(3)]
    for i in range(3):
        for j in range(9):
            list1[i].append(sudoku[a+i][j])
    for i in range(3):
        for j in range(9):
            sudoku[a+i][j]=sudoku[b+i][j]
    for i in range(3):
        for j in range(9):
            sudoku[b+i][j]=list1[i][j]

def mixSudoku4(sudoku):                      # 서로 다른 묶음 끼리의 세로줄 교환
    a, b = map(lambda x:x*3,randomTwo(0,2))
    list1=[[] for i in range(9)]
    for i in range(9):
        for j in range(3):
            list1[i].append(sudoku[i][a+j])
    for i in range(9):
        for j in range(3):
            sudoku[i][a+j]=sudoku[i][b+j]
    for i in range(9):
        for j in range(3):
            sudoku[i][b+j]=list1[i][j]

def copySudoku(sudoku, sudokuCase):          # 스도쿠를 복사하는 함수
    for i in range(9):
        for j in range(9):
            sudokuCase[i][j]=sudoku[i][j]


def makeBlankSudoku(sudoku):                 # 스도쿠에 빈칸을 만들어주는 함수
    for i in range(9):
        list1=[]
        randomIntList(0, 8, 1, list1)
        for j in list1:
            sudoku[i][j] = "0"
    return sudoku
            
def printSudoku(list1,current):              # 스도쿠를 프린트하는 함수
    print("\n|=======|=======|=======|")
    for i in range(9):
        print("| ",end='')
        for j in range(9):
            if(i==current[0]) and (j==current[1]):
                type = "<"
            else:
                type = " "
            colorSudoku(list1[i][j], type)
            if j%3==2:
                print("| ",end='')
        print()
        if i%3==2:
            print("|=======|=======|=======|")
            
def colorSudoku(value1, type):                # 스도쿠에 색깔 넣어서 프린트하는 함수
    value2=int(value1)
    if int(value2)<=0:
        print('\033[94m'+ str(abs(value2)) + '\033[0m' + type,end='')
    else:
        print(str(abs(value2)) + type, end='')
            

def insertkeypad(current, quiz):               # 사용자로부터 키보드키 입력함수
    keypad=input("(이동: wasd, 종료: q) ")
    if keypad in ['w','a','s','d']:
        shift(current, keypad)
        return 0
    elif keypad=="q":
        return 1
    elif keypad in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
        insertNumber(current, keypad, quiz)
        return 0
    
    
def shift(current, keypad):                     # 스도쿠 현재 위치 이동 함수
    if keypad=='w':
        if(current[0]!=0):
            current[0]-=1
    elif keypad=='s':
        if(current[0]!=8):
            current[0]+=1
    elif keypad=='a':
        if(current[1]!=0):
            current[1]-=1
    elif keypad=='d':
        if(current[1]!=8):
            current[1]+=1

def insertNumber(current, keypad, quiz):        # 스도쿠에 숫자를 기입하는 함수
    num=int(keypad)
    x, y = current[0] , current[1]
    if ( int(quiz[x][y]) <= 0 ):
        if num == 1:
            quiz[x][y] = -1
        elif num == 2:
            quiz[x][y] = -2
        elif num == 3:
            quiz[x][y] = -3
        elif num == 4:
            quiz[x][y] = -4
        elif num == 5:
            quiz[x][y] = -5
        elif num == 6:
            quiz[x][y] = -6
        elif num == 7:
            quiz[x][y] = -7
        elif num == 8:
            quiz[x][y] = -8
        elif num == 9:
            quiz[x][y] = -9

        
        

PlayingSudoku()
```
